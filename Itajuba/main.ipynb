```python
import psycopg2
from psycopg2 import extras
from imoveis.area_especial import area_especial 
from config import db_config
from utils.deletaPublic import clear_public_geoms
from utils.truncateTables  import truncar_tabelas_dado_novo, truncar_tabelas_to_review
from layers.eixo import eixos
from layers.logradouro import logradouros
from imoveis.processandoRotulos import carregar_lote_e_dependencias 
from layers.quadra import quadras
from utils.validacoes import validacoes_pos_carga, validacoes_pre_carga

lotesNovos = []
lotesToReviewNovos = []

def exibir_menu():
    print("\nEscolha uma opcao:")
    print("1. Validacões pré-carga")
    print("2. Processar logradouros")
    print("3. Processar eixos")
    print("4. Processar quadras")
    print("5. Carregar lote e dependências")
    print("6. Processar áreas especiais")
    print("7. Validacões pos-carga")
    print("8. Limpar geometrias públicas")
    print("9. Truncar tabelas dado_novo")
    print("10. Truncar tabelas to_review")
    print("0. Sair")
    return input("Digite o número da opcao desejada: ")

def connectDB():
    setor_carga = input("Nome da carga: ")
    # set_setor_carga(setor_carga)
    # Aqui utilizamos as configuracões importadas
    con = psycopg2.connect(host=db_config["host"], 
                           database=db_config["database"], 
                           user=db_config["user"], 
                           password=db_config["password"])
    

    while True:
        
        cur = con.cursor(cursor_factory=psycopg2.extras.DictCursor)
        opcao = exibir_menu()
        if opcao == "1":
            print("-------Carga para Produção-------")
            # validacoes_pre_carga(cur)
            
            con2 = psycopg2.connect(host=db_config["host"], 
                                database=db_config["database"], 
                                user=db_config["user"], 
                                password=db_config["password"])
            
            print("============ LOGRADOUROS =============")
            cur = con2.cursor(cursor_factory=psycopg2.extras.DictCursor)
            logradouros(cur)
            cur.close() 
            con2.commit()

            print("============ EIXOS =============")
            cur = con2.cursor(cursor_factory=psycopg2.extras.DictCursor)
            eixos(cur)
            cur.close() 
            con2.commit()

            print("============ QUADRAS =============")
            cur = con2.cursor(cursor_factory=psycopg2.extras.DictCursor)
            quadras(cur)
            cur.close() 
            con2.commit()

            print("============ LOTES =============")
            cur = con2.cursor(cursor_factory=psycopg2.extras.DictCursor)
            carregar_lote_e_dependencias(con2, cur, setor_carga)
            cur.close() 
            con2.commit()
        elif opcao == "2":
            print("============ LOGRADOUROS =============")
            logradouros(cur)
            cur.close() 
            con2.commit()
        elif opcao == "3":
            print("============ EIXOS =============")
            eixos(cur)
            cur.close() 
            con.commit()
        elif opcao == "4":
            print("============ QUADRAS =============")
            quadras(cur)
            cur.close() 
            con.commit()
        elif opcao == "5":
            carregar_lote_e_dependencias(con, cur, setor_carga)
            cur.close() 
            con.commit()
        elif opcao == "6":
            print("============ AREAS ESPECIAIS ============")
            area_especial(cur)
            cur.close() 
            con.commit()
        elif opcao == "7":
            print("-------Validacões pos carga-------")
            validacoes_pos_carga(cur)
            cur.close() 
            con.commit()
        elif opcao == "8":
            print("-------Limpando geometrias públicas-------")
            clear_public_geoms(cur)
            cur.close() 
            con.commit()
        elif opcao == "9":
            print("-------Truncando tabelas dado_novo-------")
            truncar_tabelas_dado_novo(con, cur)
            cur.close() 
            con.commit()   
        elif opcao == "10":
            print("-------Truncando tabelas to_review-------")
            truncar_tabelas_to_review(con, cur)
            cur.close() 
            con.commit()
        elif opcao == "0":
            break
        else:
            print("Opcao inválida. Por favor, tente novamente.")

    con.commit()
    con.close()
    
connectDB()
```
